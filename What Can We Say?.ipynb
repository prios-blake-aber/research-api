### What Could We Say?

```python
# ENTITY 
# MAKES AN ASSERTION ABOUT
# ENTITY
# ON MEASURE 
# FOR CONTEXT(s)
```

```python
ENTITIES = ['ENTITY - PERSON', 'ENTITY - COLLECTION OF PEOPLE', 'ENTITY - SYSTEM']
```

```python
ASSERTIONS = [
    'ASSERTION - ASSESSMENT - DOT', 
    'ASSERTION - ASSESSMENT - RESPONSE', 
#     'ASSERTION - ASSESSMENT - RANKING', 
#     'ASSERTION - ASSESSMENT - REVIEW', 
#     'ASSERTION - ASSESSMENT - TEST',
#     'ASSERTION - ESTIMATION - RESUME', 
    'ASSERTION - ESTIMATION - STREAM',
#     'ASSERTION - INTERVENTION - POPUP', 
#     'ASSERTION - INTERVENTION - TRAINING',
#     'ASSERTION - REWARD - EXTRINSIC', 
#     'ASSERTION - REWARD - INTRINSIC'
]
```

```python
CONTEXTS = [
#     'CONTEXT - LOCATION', 
    'CONTEXT - MEETING', 
#     'CONTEXT - PROJECT', 
#     'CONTEXT - TAXONOMY', 
#     'CONTEXT - TEAM', 
#     'CONTEXT - TIMEWINDOW',
#     'CONTEXT - GOAL - RESPONSIBILITY - TASK'
]
```

```python
MEASURES = [
    'MEASURE - QUALITY - ATTRIBUTE', 
#     'MEASURE - QUALITY - INTEREST', 
#     'MEASURE - QUALITY - SKILL', 
#     'MEASURE - QUALITY - VALUE',
    'MEASURE - BELIEVABILITY',
#     'MEASURE - OUTCOME'
]
```

```python
POSSIBLE_STATEMENTS_IN_GRAMMAR = []
for author, subject in itertools.product(ENTITIES, ENTITIES):
    for assertion in ASSERTIONS:
        for measure in MEASURES:
            for context in CONTEXTS:
                POSSIBLE_STATEMENTS_IN_GRAMMAR.append([
                    author, assertion, subject, measure, context
                ])
```

```python
print('Total Combinations:', len(POSSIBLE_STATEMENTS_IN_GRAMMAR))
# POSSIBLE_STATEMENTS_IN_GRAMMAR[:10]
```
